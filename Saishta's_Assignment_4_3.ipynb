# Program 1
### 1. You are given a dataset, which is present in the LMS, containing the number of hurricanes occurring in the United States along the coast of the Atlantic. Load the data from the dataset into your program and plot a Bar Graph of the data, taking the Year as the x-axis and the number of hurricanes occurring as the Y-axis.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Reading the data.
data = pd.read_csv("Hurricanes.csv")
data.head()
```

```python
# Plotting a bar graph.
plt.figure(figsize=(8,8))
plt.bar(data["Year"],data["Hurricanes"])
plt.xlabel("Year")
plt.ylabel("Number of Hurricanes")
plt.title("Number of hurricanes occurring in the United States along the coast of the Atlantic")
plt.show
```

# Program 2
### 2. The dataset given, records data of city temperatures over the years’ 2014 and 2015. Plot the histogram of the temperatures over this period for the cities of San Francisco and Moscow.

```python
data2 = pd.read_csv("CityTemps.csv")
data2.head()
```

```python
plt.figure(figsize=(8,8))
plt.hist(data2["San Francisco"], label = "San Francisco", bins=15, alpha=0.5)
plt.hist(data2["Moscow"], label="Moscow", bins=15,alpha=0.5)
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.title("Histogram of the temperatures over this period for the cities of San Francisco and Moscow.")
plt.legend()
plt.show()
```

# Program 4
### Let the x axis data points and y axis data points are
### X = [1,2,3,4]
### y = [20, 21, 20.5, 20.8]

```python
# Drawing a simple plot.
x = [1,2,3,4]
y = [20, 21, 20.5, 20.8]
plt.figure(figsize=(5,5))
plt.plot(x,y)
plt.show()
```

```python
# Configuring the line and markers in simple plot.
plt.plot(x,y,linestyle= "--", marker="o")
plt.show()
```

```python
# Configuring the axis.
plt.plot(x,y,linestyle= "--", marker="o")
plt.xlim(0,5)
plt.ylim(19.5,21.5)
plt.show()
```

```python
# Giving title of graph and labels of x axis and yaxis.
plt.plot(x,y,linestyle= "--", marker="o")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Simple Plot")
```

```python
# Giving error bar if y_error = [0.12, 0.13, 0.2, 0.1]
y_error = [0.12, 0.13, 0.2, 0.1]
plt.errorbar(x,y,yerr = y_error)
plt.show()
```

```python
# define width, height as figsize=(4,5) DPI and adjust plot dpi=100
plt.figure(figsize=(4,5), dpi=100)
plt.plot(x,y,linestyle= "--", marker="o")
plt.show()
```

```python
# Give a font size of 14.
plt.plot(x,y,linestyle= "--", marker="o")
plt.rc("font",size=14)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Simple Plot")
plt.show()
```

```python
# Create a dataframe.
givendata= {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
'female': [0, 1, 1, 0, 1],
'age': [42, 52, 36, 24, 73],
'preTestScore': [4, 24, 31, 2, 3],
'postTestScore': [25, 94, 57, 62, 70]}
df = pd.DataFrame(givendata)
df.head()
```

```python
# Draw a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age.
plt.figure(figsize=(5,5))
plt.scatter(df["preTestScore"],df["postTestScore"], s=df["age"],alpha=0.5)
plt.xlabel("Pre test score")
plt.ylabel("Post test score")
plt.title("Scatter Plot")
plt.show()
```

```python
# Draw a Scatterplot from the data in question 9 of preTestScore and postTestScore with the size = 300 and the color determined by sex.
plt.figure(figsize=(5,5))
colors = ["blue" if g==0 else "red" for g in df["female"]]
plt.scatter(df["preTestScore"],df["postTestScore"],c=colors, s=300,alpha=0.5,)
plt.xlabel("Pre test score")
plt.ylabel("Post test score")
plt.title("Scatter Plot")
plt.show()
```
