```python
# Reading Data
import pandas as pd
data = pd.read_csv("FairDealCustomerData.csv")
data.head()
```

```python
# Separating full name into title, first name and last name.
def split_name(full_name):
  import re
  name_parts = re.split(r'\s+',full_name)
  title = name_parts[1] if len(name_parts)>1 else ''
  first_name = name_parts[2]
  last_name = name_parts[3]
  return title,first_name,last_name

# Applying function to the dataframe.
data["Title"],data["FirstName"],data["LastName"] = zip(*data["Name"].map(split_name))
data.head()
```

```python
# Storing data in customer class
# Define the customer class
class Customer:
  def __init__ (self, customer_id,title,first_name,last_name,blacklisted):
    self.customer_id = customer_id
    self.title = title
    self.first_name = first_name
    self.last_name = last_name
    self.blacklisted = blacklisted

# creting a list of customer objects
customers = []
for index, row in data.iterrows():
  customer = Customer(index,row["Title"],row["FirstName"],row["LastName"], row["Blacklisted"])
  customers.append(customer)

# Displaying few customer objects
for i in range(5):
  print(customers[i].__dict__)
```

```python
# Create Custom Exception - CustomerNot AllowedException
# Defining custom exception class
class CustomerNotAllowedException(Exception):
  def __init__ (self,message):
    super().__init__(message)

# Defining the create order function.
class Order:
  def __init__(self,customer_id,first_name,last_name, product_name, product_code):
    self.customer_id = customer_id
    self.first_name = first_name
    self.last_name = last_name
    self.product_name = product_name
    self.product_code = product_code

def create_order(customer,product_name,product_code):
      if customer.blacklisted == 1:
        raise CustomerNotAllowedException(f"Customer {customer.customer_id} is not allowed to create an order.")
      else:
        order = Order(customer.customer_id, customer.first_name,customer.last_name, product_name, product_code)
        return order

# Example usage:
try:
  order = create_order(customers[0], "Sample Product", "P123")
  print("Order created successfully: ", order.__dict__)
except CustomerNotAllowedException as e:
  print("Error:",e)
```
