# Program 1

### Load the data from “cereal.csv” and plot histograms of sugar and vitamin content across different cereals.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Loding the data
data = pd.read_csv("cereal.csv")
data.head()
```

```python
data.columns
```

```python
# Separating the columns.
x = data["sugars"]
y= data["vitamins"]
```

```python
# Histogram for sugar.
plt.figure(figsize=(8,8))
plt.hist(x, bins=20, edgecolor ="k")
plt.title("Histogram of Sugar content")
plt.xlabel("Sugar Content")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plotting Histogram for Vitamins.
plt.subplot(1,2,2)
plt.hist(y, bins=20, edgecolor ="r")
plt.title("Histogram of Vitamin content")
plt.xlabel("Vitamin Content")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

# Program 2

### Create a bar plot where each manufacturer is on the y axis and the height of the bars depict the number of cereals manufactured by them.

```python
data.columns
```

```python
full_name = {'N': 'Nabisco',
'Q': 'Quaker Oats',
'K': 'Kelloggs',
'R': 'Raslston Purina',
'G': 'General Mills' ,
'P' :'Post' ,
'A':'American Home Foods Products'}
data["Manufacturer_Full"] = data["mfr"].map(full_name)
data.head()
```

```python
import seaborn as sns
plt.figure(figsize=(8,8))
sns.countplot(y = "Manufacturer_Full", data = data, order= data["Manufacturer_Full"].value_counts().index)
plt.title("No. of cereals manufactured by each manufacturer")
plt.xlabel("Number of Cereals")
plt.ylabel("Manufacturer")
plt.show()
```

# Program 3

### 3. Extract the rating as your target variable ‘y’ and all numerical parameters as your predictors ‘x’. Separate 25% of your data as test set.

```python
data.columns
```

```python
from sklearn.model_selection import train_test_split

x = data.select_dtypes(include="number")
y = data["rating"]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
```

# Program 4

### 4. Fit a linear regression module and measure the mean squared error on test dataset.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model=LinearRegression()
model.fit(x_train,y_train)

y_pred=model.predict(x_test)

mse =mean_squared_error(y_test,y_pred)
mse
```
