# Program 1.
## Load the file “zoo.data” and look at the info and first five rows. The first column denotes the animal name and the last one specifies a high-level class for the corresponding animal.

```python
import pandas as pd
data = pd.read_csv("zoo.csv")
data.head()
```

```python
data.info()
```

# Program 2
## Find out the unique number of high level class.

```python
unique_number= data["class_type"].unique()
unique_number
```

# Program 3
## Use the 16-intermediate feature and perform an agglomerative clustering.
## [ Hint: Refer to the agglomerative clustering module in scikit learn and set the number of clusters appropriately]

```python
import numpy as np
from sklearn.cluster import AgglomerativeClustering
x = data.iloc[ : , 1:17].to_numpy()
n_clusters = 3
acm = AgglomerativeClustering(n_clusters=n_clusters)
clusters = acm.fit_predict(x)
```

```python
print("Cluster assignmnets: ")
for animal,cluster in zip(data["animal_name"],clusters):
  print(f"{animal}: Cluster {clusters}")
```

# Program 4
## Compute the mean squared error by comparing the actual class and predicted high level class.

```python
from sklearn.metrics import mean_squared_error
actual_class=data["class_type"].values
mse = mean_squared_error(actual_class, clusters)
print("Mean Squared Error: ",mse)
```
