```python
# Loading the data.
import pandas as pd
data = pd.read_csv("loan_borowwer_data.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.columns
```

```python
# Converting "purpose" into numerical format.
data= pd.get_dummies(data,columns=["purpose"],drop_first=True)
data.head()
```

```python
data.columns
```

```python
# Finding the correlation and creating a heatmap.
import seaborn as sns
import matplotlib.pyplot as plt
correlation = data.corr()

plt.figure(figsize=(15,10))
sns.heatmap(correlation, cmap="coolwarm",annot= True)
plt.title("Correlation heatmap")
plt.show()
```

```python
# Splitting the data.
x = data.drop(columns=["not.fully.paid"])
y= data["not.fully.paid"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=2)
```

```python
# Appling Linear Regression model.
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
# Evaluating our model error.
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print("The mean squared error is: ",mse)
print("The r2 score is: ",r2)
```

```python
# Coefficient and their corresponding features.
coeff = pd.DataFrame({"Feature": x.columns,"Coefficient":lr.coef_})
coeff
```
