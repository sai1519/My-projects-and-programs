```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Loading the data
data =pd.read_csv("HollywoodMovies.csv")
data.head()
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
# Handeling the missing data.
valid_data=data.dropna(subset=["Story","RottenTomatoes"])
```

```python
valid_data.isnull().sum()
```

# Find the highest rated movie in the “Quest” story type.

```python
questmovie = valid_data[valid_data["Story"]=="Quest"]
highest_rating = questmovie.loc[questmovie["RottenTomatoes"].idxmax()]
print("Highest rated movie in the Quest story type: ")
print(highest_rating)
```

# Find the genre in which there has been the greatest number of movie releases

```python
genre_count = data["Genre"].value_counts()
most_common_genre=genre_count.idxmax()
print("Genre with the greatest number of movie releases: ",most_common_genre)
```

# Print the names of the top five movies with the costliest budgets.

```python
top_five= data.nlargest(5,"Budget")
print("Top 5 movies with the costliest budgets: ", top_five[["Movie","Budget"]])
```

# Is there any correspondence between the critics’ evaluation of a movie and its acceptance by the public? Find out, by plotting the net profitability of a movie against the ratings it receives on Rotten Tomatoes.

```python
data.columns
```

```python
plt.figure(figsize=(5,5))
plt.scatter(data["RottenTomatoes"],data["Profitability"])
plt.xlabel("Rotten Tomatoes Rating")
plt.ylabel("Profitability")
plt.title("Net profitability vs Rotten Tomatoes")
plt.show()
```

### 6.10: Create pandas dataframe having keys and ltable and rtable as below -
### 'key': ['One', 'Two'], 'ltable': [1, 2]
### 'key': ['One', 'Two'], 'rtable': [4, 5]
### Merge both the tables based of key

```python
givendatal = {'key': ['One', 'Two'], 'ltable': [1, 2]}
givendatar={'key': ['One', 'Two'], 'rtable': [4, 5]}
df10 = pd.DataFrame(givendatal)
df11 =pd.DataFrame(givendatar)
mergeddf12=pd.merge(df10,df11,on="key")
mergeddf12
```
