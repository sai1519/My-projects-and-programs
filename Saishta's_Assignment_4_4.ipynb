```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Reading the data.
data =pd.read_csv("BigMartSalesData.csv")
data.head()
```

```python
data.columns
```

```python
# Plotting total sales per month for year 2011.
# Ensuring the "Invoicedate" column is in the correct date format.
data["InvoiceDate"]=pd.to_datetime(data["InvoiceDate"],format="%d-%m-%y")
#Extracting the year and month from the invoicedate column.
data["Year"]=data["InvoiceDate"].dt.year
data["Month"]=data["InvoiceDate"].dt.month
data.head()
```

```python
# Filter data for year 2011.
year2011 = data[data["Year"]==2011]
# Group by month and calculated total sales fr each month.
monthlysales = year2011.groupby("Month")["Amount"].sum()
year2011.head()
```

```python
monthlysales
```

```python
# Plotting the data.
plt.figure(figsize=(5,5))
plt.plot(monthlysales.index,monthlysales.values,marker="o")
plt.title("Total sale Per Month for Year 2011")
plt.xlabel("month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
# Finding the month with the lowest sales.
lowestsalemonth= monthlysales.idxmin()
lowestsalevalue = monthlysales.min()
print(f"The Month with the lowest sales in 2011 is: {lowestsalemonth}. With total sale value: {lowestsalevalue}")
```

```python
# Plotting data in bar chart.
plt.figure(figsize=(5,5))
plt.bar(monthlysales.index,monthlysales.values)
plt.title("Total sales per Month for year 2011 (Bar Chart)")
plt.xlabel("Month")
plt.ylabel('Total Sales')
plt.grid(axis="y")
plt.show()
```

```python
year2011.columns
```

```python
# Plotting country wise pie chart.
# Grouping by country and calculating total sales for each country in 2011.
countrysales = year2011.groupby("Country")["Amount"].sum()
countrysales
```

```python
# Plotting data in pie chart.
plt.figure(figsize=(8,8))
plt.pie(countrysales,labels=countrysales.index,autopct="%1.1f%%",shadow=True,startangle=90)
plt.title("Year 2011 sales by country")
plt.show()
```

```python
# Country with highest sale.
hscountry = countrysales.idxmax()
hscountryvalue = countrysales.max()
print(f"Highest contributing contry: {hscountry} with {hscountryvalue} sale")
```

```python
# Plotting a scatter plot for the invoice amounts.
plt.figure(figsize=(5,5))
plt.scatter(data["InvoiceNo"],data["Amount"], color="red",alpha=0.5)
plt.title("Scatter plot of invoice Amounts")
plt.xlabel("Invoice Number")
plt.ylabel("Invoice Amount")
plt.grid(True)
plt.show()
```
