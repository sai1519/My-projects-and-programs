```python
# Reading the data.
import numpy as np
import pandas as pd
mathdata = pd.read_csv("MathScoreTerm1.csv")
physicsdata = pd.read_csv("PhysicsScoreTerm1.csv")
dsdata=pd.read_csv("DSScoreTerm1.csv")
print(mathdata.head())
print(physicsdata.head())
print(dsdata.head())
```

```python
print(mathdata.columns)
print(physicsdata.columns)
print(dsdata.columns)
```

```python
# Removing Name and ethinicity columns.
x = mathdata.drop(['Name','Ethinicity'],axis =1)
y = physicsdata.drop(['Name','Ethinicity'], axis=1)
z = dsdata.drop(['Name','Ethinicity'],axis = 1)
```

```python
# Filling the missing score data with zero.
x.fillna(0,inplace=True)
y.fillna(0,inplace =True)
z.fillna(0,inplace=True)
```

```python
# Merging the 3 files.
merged = pd.merge(x,y, on = "ID",how="inner")
merged = pd.merge(merged,z, on = "ID",how="inner")
merged.head()
```

```python
# Changing Sex M/F to 1/2
merged["Sex"] =merged["Sex"].map({"M":1,"F":2})
merged.head()
```

```python
# Removing redundent data.
merged.drop(["Age_x","Sex_x","Age_y","Sex_y"],axis =1, inplace=True)
merged.head()
```

```python
# Storing the data in a new file.
merged.to_csv("ScoreFinal.csv",index = False)
```

```python
# Converting ethnicity to numerical value
merged2 = pd.merge(mathdata,physicsdata, on="Name", how="inner")
merged2 = pd.merge(merged2,dsdata,on="Name",how="inner")
merged2.drop(["Age_x","Ethinicity_x","Sex_x","ID_x","Age_y","Ethinicity_y","Sex_y","ID_y"],axis = 1, inplace=True)
merged2.head()
```

```python
merged2["Ethinicity"].unique()
```

```python
emapping = {"White American": 1, "European American": 2,"Hispanic": 3,"African American": 4}
merged2["Ethinicity"]=merged2["Ethinicity"].str.strip().str.lower()
default_value = -1
merged2["Ethinicity"]=merged2["Ethinicity"].map(emapping)
merged2.head()
```

```python
# Filling the NaN values with mean.
merged2["Score_x"].fillna(merged2['Score_x'].mean(), inplace =True)
merged2["Score_y"].fillna(merged2['Score_y'].mean(), inplace =True)
merged2["Score"].fillna(merged2['Score'].mean(), inplace =True)
```

```python
# Saving the updated data frame in the new file.
merged2.to_csv("ScoreFinal_WithEnhancements.csv",index = False)
```
