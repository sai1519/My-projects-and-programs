# 1) Extract data from the given SalaryGender CSV file and store the data from each column in a separate NumPy array

```python
# Reading the csv file.
import numpy as np
import pandas as pd
data = pd.read_csv("SalaryGender.csv")
data.head()
```

```python
# Storing data in separate numpy array.
salary = np.array(data["Salary"])
gender = np.array(data["Gender"])
age = np.array(data["Age"])
phd = np.array(data["PhD"])
print("Salary is: \n", salary)
print("Gender is: \n", gender)
print("Age is: \n", age)
print("PhD is : \n", phd)
```

# 2) Find:
# 1. The number of men with a PhD
# 2. The number of women with a PhD

```python
# Finding no. of men with phd.
phdmen = np.sum((gender == 1) & (phd == 1))
phdmen
```

```python
# Finding no. of women with phd.
phdwomen = np.sum((gender == 0) & (phd == 1))
phdwomen
```

# 5) How do you Count The Number Of Times Each Value Appears In An Array Of Integers?
# [0, 5, 4, 0, 4, 4, 3, 0, 0, 5, 2, 1, 1, 9]
# Answer should be array([4, 2, 1, 1, 3, 2, 0, 0, 0, 1]) which means 0 comes 4 times,
# 1 comes 2 times, 2 comes 1 time, 3 comes 1 time and so on.

```python
arr = np.array([0, 5, 4, 0, 4, 4, 3, 0, 0, 5, 2, 1, 1, 9])
count = np.bincount(arr) # Value_counts is the function of pandas and not of numpy.
count
```

# 7. Create a numpy array having NaN (Not a Number) and print it.
# array([ nan, 1., 2., nan, 3., 4., 5.])Print the same array omitting all elements which are nan

```python
arr2 = np.array([ np.nan, 1., 2., np.nan, 3., 4., 5.]) # np.nan is the standard way to write nan value in numpy.
arr3 = arr2[~np.isnan(arr2)] # ~ is for not.
arr3
```

# 13. Create a random array and swap two rows of an array.

```python
np.random.seed(1)
arr6 = np.random.random(size=(5,5))
print("The original array is: \n",arr6)
row1,row2 = 1,4
arr6[[row1,row2]]= arr6[[row2,row1]]
print("The swaped array is: \n", arr6)
```

# 14. Create a random matrix and Compute a matrix rank.

```python
arr7 = np.random.random(size=(3,3))
print("The array is: \n",arr7)
rank = np.linalg.matrix_rank(arr7)
print("The rank of the matrix is: \n", rank)
```
