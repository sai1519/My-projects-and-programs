# Program 2

### a. Create a new column -’total_benefitted’ that is a sum of inmates benefitted through all modes.

```python
data["total_benefitted"]=data[["No. of Inmates benefitted by Elementary Education","No. of Inmates benefitted by Adult Education","No. of Inmates benefitted by Higher Education","No. of Inmates benefitted by Computer Course"  ]].sum(axis=1)
data.head()
```

### Create a new row - “totals” that is the sum of all inmates benefitted through each mode across all states.

```python
total =data[["No. of Inmates benefitted by Elementary Education","No. of Inmates benefitted by Adult Education","No. of Inmates benefitted by Higher Education","No. of Inmates benefitted by Computer Course"  ]].sum()
newdf = pd.DataFrame(total).transpose()
newdf.index=["totals"]
data=data.append(newdf,ignore_index=False)
data.tail()
```

# Program 3

### a. Make a bar plot with each state name on the x -axis and their total benefitted inmates as their bar heights. Which state has the maximum number of beneficiaries?

```python
import matplotlib.pyplot as plt
states = data[data.index!="totals"]
x = states["STATE/UT"]
y=states["total_benefitted"]
plt.figure(figsize=(6,6))
plt.bar(x,y)
plt.xlabel("States")
plt.ylabel("Total Benefitted")
plt.title("Bar Graph to know state with max beneficiaries")
plt.xticks(rotation=90)
mbs = x[y.idxmax()]
print(f"The state with the maximum number of beneficiary is: {mbs}")
plt.show()
```

### b. Make a pie chart that depicts the ratio among different modes of benefits.

```python
data.columns
```

```python
benefit_mode = ["No. of Inmates benefitted by Elementary Education","No. of Inmates benefitted by Adult Education","No. of Inmates benefitted by Higher Education","No. of Inmates benefitted by Computer Course"]
benefit_count = states[benefit_mode].sum()
benefit_count
```

```python
plt.figure(figsize=(8,8))
plt.pie(benefit_count, labels=benefit_count.index,autopct="%1.1f%%", startangle=140)
plt.axis("equal")
plt.title("Ratio of differnt Modes of Benefits")
plt.show()
```
